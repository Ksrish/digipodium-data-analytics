load the data analysis libraries

```python
import numpy as np #numbers used
import matplotlib.pyplot as plt 
```

get the data/ generate data

```python
rainfall = [17,10,10,10,26,59]
months = ['jan','feb','mar','apr','may','jun']
```

visualize the data

```python
plt.bar(months,rainfall,color ='#ff7700')
```

```python
plt.style.available
```

```python
plt.style.use('ggplot') #theme
```

```python
plt.bar(months, rainfall)
```

```python
plt.pie(
    
    rainfall,
    labels = months,
    autopct='%1.1f%%',

)
plt.show()
```

```python
plt.pie(
    
    rainfall,
    labels = months,
    autopct='%.2f%%',

)
plt.show()
```

```python
plt.pie(
    
    rainfall,
    labels = months,
    autopct='%2.3f%%',

)
plt.show()
```

```python
plt.pie(
    
    rainfall,
    labels = months,
    autopct='%1.1f%%',
    wedgeprops={'width':0.4,'edgecolor':'#666611','linewidth':2},
    pctdistance=.8

)
plt.show()
```

```python
plt.boxplot(rainfall,
            vert = False,
            showmeans=True,
            meanline=True,
            meanprops={'color':'red'}

            )
plt.show()
```
